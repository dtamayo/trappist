```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
plt.style.use('paper')

fig = plt.figure(figsize=(15,15))
gs = gridspec.GridSpec(2, 2)
gs.update(wspace=0.3)

ax0 = plt.subplot(gs[0, 0])
ax1 = plt.subplot(gs[1, 0])
ax2 = plt.subplot(gs[:, 1])


cm = plt.cm.get_cmap('RdYlBu_r')
dfstable = df[df['Eerrf']<1.e-2]
dfshort = dfstable[dfstable['tinstability']<=5.e6]
dflong = dfstable[dfstable['tinstability']>5.e6]
sc = ax0.scatter(dfshort['K'], dfshort['mag'], c=dfshort['max1storder3MMRlibamp1'], edgecolor='', vmin=0, vmax=180, s=80,cmap=cm)        
sc = ax0.scatter(dflong['K'], dflong['mag'], c=dflong['max1storder3MMRlibamp1'], edgecolor='', vmin=0, vmax=180, s=320,cmap=cm)        
sc = ax1.scatter(dfshort['K'], dfshort['mag'], c=dfshort['max3MMRlibamp1'], edgecolor='', vmin=0, vmax=180, s=80,cmap=cm)        
sc = ax1.scatter(dflong['K'], dflong['mag'], c=dflong['max3MMRlibamp1'], edgecolor='', vmin=0, vmax=180, s=320,cmap=cm)        
dfunstable = df[df['Eerrf']>1.e-2]
dfshort = dfunstable[dfunstable['tinstability']<=5.e6]
dflong = dfunstable[dfunstable['tinstability']>5.e6]
sc = ax0.scatter(dfshort['K'], dfshort['mag'], c=dfshort['max1storder3MMRlibamp1'], edgecolor='', vmin=0, vmax=180, s=200,cmap=cm, marker='x')        
sc = ax0.scatter(dflong['K'], dflong['mag'], c=dflong['max1storder3MMRlibamp1'], edgecolor='', vmin=0, vmax=180, s=400,cmap=cm, marker='x')        
sc = ax1.scatter(dfshort['K'], dfshort['mag'], c=dfshort['max3MMRlibamp1'], edgecolor='', vmin=0, vmax=180, s=200,cmap=cm, marker='x')        
sc = ax1.scatter(dflong['K'], dflong['mag'], c=dflong['max3MMRlibamp1'], edgecolor='', vmin=0, vmax=180, s=400,cmap=cm, marker='x')        

ax0.set_xlabel(r'$K$')
ax1.set_xlabel(r'$K$')
ax0.set_xscale('log')
ax0.set_yscale('log')
ax1.set_xscale('log')
ax1.set_yscale('log')
ax0.set_xlim([10,1000])
ax1.set_xlim([10,1000])
ax0.set_ylim([1.e-3, 1])
ax1.set_ylim([1.e-3, 1])

ax0.set_ylabel(r'$\eta$')  
ax1.set_ylabel(r'$\eta$')  
ax0.set_title('Max Libration Amplitude (1st order MMRs)')
ax1.set_title('Max Libration Amplitude (all MMRs)')

cbar_ax = fig.add_axes([0.50, 0.15, 0.01, 0.7])
fig.colorbar(sc, cax=cbar_ax)

P_obs = {'b':1.51087081, 'c':2.4218233, 'd':4.049610, 'e':6.099615, 'f':9.206690, 'g':12.35294, 'h':18.764} # days
Pratio_obs = {pair:P_obs[pair[1]]/P_obs[pair[0]] for pair in resonances.keys()}

xlim = 0.02
Nbins=80

# first get ylim
maxbintotal=0
for i, resonance in enumerate(resonances.items()):
    pair = resonance[0]
    res = resonance[1]
    resratio = res[0]/res[1]
    
    bins = np.linspace(0, xlim, Nbins, endpoint=True)
    hist, bins = np.histogram(df['Pratio1'+str(pair[0])+str(pair[1])]-resratio, bins=bins)
    maxbintotal+=hist.max()
height = maxbintotal

maxbintotal=0
for i, resonance in enumerate(resonances.items()):
    pair = resonance[0]
    res = resonance[1]
    resratio = res[0]/res[1]
    
    bins = np.linspace(0, xlim, Nbins, endpoint=True)
    hist, bins = np.histogram(df['Pratio1'+str(pair[0])+str(pair[1])]-resratio, bins=bins)
    center = (bins[:-1] + bins[1:]) / 2
    bar = ax2.bar(center, hist, align='center', width=np.diff(bins), bottom=maxbintotal, label=pair)
    
    ax2.axvline(Pratio_obs[pair]-resratio, ymin=maxbintotal/height, ymax = (maxbintotal + hist.max())/height, color=bar.patches[0].get_facecolor())
    ax2.text(0.014, maxbintotal + hist.max()/2, str(pair[1])+'/'+str(pair[0]), fontsize=20, color=bar.patches[0].get_facecolor())
    ax2.arrow( 0.0015, maxbintotal + hist.max()/2, 0.0035, 0., head_width=7, head_length=0.001, fc="k", ec="k", lw=3)
    maxbintotal+=hist.max()
    
ax2.set_xlim([0,xlim])
ax2.set_ylim([0,height])
ax2.yaxis.set_visible(False)
ax2.set_xlabel('Period ratio deviation from nominal resonance')
ax2.set_ylabel('Normalized number of simulations')
ax2.text(0.003, 155, 'Tides', fontsize=20)
plt.savefig("IC.pdf")
```
